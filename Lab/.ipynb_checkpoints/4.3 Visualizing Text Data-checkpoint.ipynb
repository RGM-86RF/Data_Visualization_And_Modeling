```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
import string
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS as sklearn_stop_words
import plotly.express as px

df = pd.read_csv('IMDB_Dataset_Cleaned.csv')
df.head()
```

```python


# Define additional stopwords to remove
additional_stopwords = { 'really', 'think' , 'look', 'way', 'good'}

# Separate positive and negative reviews
positive_reviews = ' '.join(df[df['sentiment'] == 'positive']['review'].astype(str))
negative_reviews = ' '.join(df[df['sentiment'] == 'negative']['review'].astype(str))

# Tokenize the words
postive_words = positive_reviews.lower().split()
    
# Remove punctuation and stopwords (including additional stopwords)
postive_words = [word.strip(string.punctuation) for word in postive_words if word.isalnum() and word not in sklearn_stop_words and word not in additional_stopwords]
    
# Generate the word cloud
print('Positive Reviews')
wordcloud = WordCloud(width=800, height=400, background_color ='white').generate(' '.join(postive_words))
plt.figure(figsize=(10, 7)) 
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

#Negative reviews
negative_words = negative_reviews.lower().split()
    
# Remove punctuation and stopwords (including additional stopwords)
negative_words = [word.strip(string.punctuation) for word in negative_words if word.isalnum() and word not in sklearn_stop_words and word not in additional_stopwords]
    
# Generate the word cloud
print('Negative Reviews')
wordcloud = WordCloud(width=800, height=400, background_color ='white').generate(' '.join(negative_words))
plt.figure(figsize=(10, 7)) 
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

The figures above is a word cloud that shows the most popular words used in positive and negative movie reviews on imdb. The larger the word is shows how frequent they are used in reviews.

```python
additional_stopwords = { 'really', 'think' , 'look', 'way'}

# Separate positive and negative reviews
positive_reviews = ' '.join(df[df['sentiment'] == 'positive']['review'].astype(str))
negative_reviews = ' '.join(df[df['sentiment'] == 'negative']['review'].astype(str))

# Tokenize the words
postive_words = positive_reviews.lower().split()
    
# Remove punctuation and stopwords (including additional stopwords)
postive_words = [word.strip(string.punctuation) for word in postive_words if word.isalnum() and word not in sklearn_stop_words and word not in additional_stopwords]

negative_words = negative_reviews.lower().split()
    
# Remove punctuation and stopwords (including additional stopwords)
negative_words = [word.strip(string.punctuation) for word in negative_words if word.isalnum() and word not in sklearn_stop_words and word not in additional_stopwords]


positive_counts = Counter(postive_words)
negative_counts = Counter(negative_words)

top_pos = positive_counts.most_common(15)
top_neg = negative_counts.most_common(15)

fig, ax = plt.subplots(1, 2, figsize=(16, 6))

# Positive bar chart
ax[0].bar([w for w, _ in top_pos], [c for _, c in top_pos], color='green')
ax[0].set_title("Top 15 Words in Positive Reviews")
ax[0].set_xticks(range(len(top_pos)))
ax[0].set_xticklabels([w for w, _ in top_pos], rotation=45)
ax[0].set_ylabel("Frequency")

# Negative bar chart
ax[1].bar([w for w, _ in top_neg], [c for _, c in top_neg], color='red')
ax[1].set_title("Top 15 Words in Negative Reviews")
ax[1].set_xticks(range(len(top_neg)))
ax[1].set_xticklabels([w for w, _ in top_neg], rotation=45)
ax[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

For this visualization I decided to showcase the top 15 words used in both positive and negative reviews. Doing this you can see what the similarites between the most used words were and the frequency of them.

```python
df['review_length'] = df['review'].str.len()

fig=px.box(df,
           x='sentiment',
           y='review_length')

fig.update_layout(title='IMDB Movie Review Text Length by Sentiment')

fig.show()
```

For this visualization you can see the differences in length between positive and negative reviews. As you can see the positive reviews tend to be longer in length compared to negative reviews.
