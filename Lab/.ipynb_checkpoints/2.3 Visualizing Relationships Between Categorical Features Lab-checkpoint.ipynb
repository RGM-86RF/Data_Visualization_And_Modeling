```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

df = pd.read_csv('titanic_passengers.csv')

df.head(20)
```

```python
df.loc[df['Fare'] < 8, 'Fare_num'] = 'Poor'
df.loc[(df['Fare'] >= 8) & (df['Fare'] < 31) , 'Fare_num'] = 'Average'
df.loc[df['Fare'] >= 31 , 'Fare_num'] = 'Rich'


df.loc[df['Survived'] == 0, 'Fare_num'].value_counts().plot(kind='pie')
df.loc[df['Survived'] == 0, 'Fare_num'].value_counts()
```

```python
df.loc[df['Fare'] < 8, 'Fare_num'] = 'Poor'
df.loc[(df['Fare'] >= 8) & (df['Fare'] < 31) , 'Fare_num'] = 'Average'
df.loc[df['Fare'] >= 31 , 'Fare_num'] = 'Rich'


df.loc[df['Survived'] == 1, 'Fare_num'].value_counts().plot(kind='pie')
df.loc[df['Survived'] == 1, 'Fare_num'].value_counts()
```

```python
df.loc[df['Age'] < 18, 'Age_num'] = 'Not Adult'
df.loc[(df['Age'] >= 18) & (df['Age'] < 35) , 'Age_num'] = 'Young Adult'
df.loc[(df['Age'] >= 30) & (df['Age'] < 60) , 'Age_num'] = 'Middle Aged'
df.loc[df['Age'] >= 60, 'Age_num'] = 'Old'

df.loc[df['Survived'] == 0, 'Age_num'].value_counts().plot(kind='pie')
df.loc[df['Survived'] == 0, 'Age_num'].value_counts()
```

```python
df.loc[df['Age'] < 18, 'Age_num'] = 'Not Adult'
df.loc[(df['Age'] >= 18) & (df['Age'] < 35) , 'Age_num'] = 'Young Adult'
df.loc[(df['Age'] >= 30) & (df['Age'] < 60) , 'Age_num'] = 'Middle Aged'
df.loc[df['Age'] >= 60, 'Age_num'] = 'Old'

df.loc[df['Survived'] == 1, 'Age_num'].value_counts().plot(kind='pie')
df.loc[df['Survived'] == 1, 'Age_num'].value_counts()
```

```python
least_df = df.loc[(df['Age'] >= 20) & (df['Age'] <= 60) & (df['Sex'] == 'male') & (df['Fare'] > 5) &
(df['Fare'] < 40) & (df['Pclass'] == 3) & (df['Embarked'] == 'S') & (df['Parch']== 0) & (df['SibSp']== 0)]
least_df.head(10)
```

```python
most_df = df.loc[(df['Age'] >= 18) & (df['Age'] <= 60) & (df['Fare'] >15) & (df['Fare'] < 130) & 
(df['Sex'] == 'female') & (df['Parch'] < 1) &(df['Embarked'] == 'S') & (df['Pclass'] == 1) & (df['SibSp']== 0)]
most_df.head(20)
```

```python
plt.figure(figsize=(8,4))
df['Age'].hist(bins=20, color='skyblue', edgecolor='black')
plt.axvline(18, color='red', linestyle='--', label='Min Age = 18')
plt.axvline(60, color='green', linestyle='--', label='Max Age = 60')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend()
plt.show()

# Fare distribution
plt.figure(figsize=(8,4))
df['Fare'].hist(bins=20, color='orange', edgecolor='black')
plt.axvline(15, color='red', linestyle='--', label='Min Fare = 15')
plt.axvline(130, color='green', linestyle='--', label='Max Fare = 130')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.legend()
plt.show()

# Sex distribution
df['Sex'].value_counts().plot(kind='bar', color='purple')
plt.title('Sex Distribution')
plt.ylabel('Count')
plt.show()

# Parch distribution
df['Parch'].value_counts().sort_index().plot(kind='bar', color='teal')
plt.title('Parents/Children Aboard (Parch)')
plt.xlabel('Parch')
plt.ylabel('Count')
plt.show()

# Embarked distribution
df['Embarked'].value_counts().plot(kind='bar', color='pink')
plt.title('Port of Embarkation')
plt.ylabel('Count')
plt.show()

# Pclass distribution
df['Pclass'].value_counts().sort_index().plot(kind='bar', color='lightgreen')
plt.title('Passenger Class Distribution')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()

# SibSp distribution
df['SibSp'].value_counts().sort_index().plot(kind='bar', color='gold')
plt.title('Siblings/Spouses Aboard (SibSp)')
plt.xlabel('SibSp')
plt.ylabel('Count')
plt.show()
```
