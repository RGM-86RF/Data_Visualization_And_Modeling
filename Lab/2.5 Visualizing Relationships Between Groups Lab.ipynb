# Lab Instructions

You are part of a data team for a financial institution that uses an AI model to determine who will and will not be granted a loan.  Recently, there have been complaints from customers that the model unfairly discriminates based on the applicant's age and sex. A court has ordered that the financial institution must provide information about how the model works.  

Data from recent loan applications is given in `loan_data.csv`.  The data dictionary is below:

| Feature              | Description | Data Type |
| :---------------- | :------: | ----: |
|person_age|Age of the person|Float
|person_gender|Gender of the person|Categorical
|person_education|Highest education level|Categorical
|person_income|Annual income|Float
|person_emp_exp|Years of employment experience|Integer
|person_home_ownership|Home ownership status (e.g., rent, own, mortgage)|Categorical
|loan_amnt|Loan amount requested|Float
|loan_intent|Purpose of the loan|Categorical
|loan_int_rate|Loan interest rate|Float
|loan_percent_income|Loan amount as a percentage of annual income|Float
|cb_person_cred_hist_length|Length of credit history in years|Float
|credit_score|Credit score of the person|Integer
|previous_loan_defaults_on_file|Indicator of previous loan defaults|Categorical
|loan_status (target variable)|Loan approval status: 1 = approved; 0 = rejected|Integer

Visualize the relationship between age, sex and 5 other features with `loan_status`.  Which features seem to drive the model's decision to grant a loan?  Which seem unrelated?  Should the financial institution be concerned about potential legal trouble?  Make sure to justify your answers with specific references to your visualizations.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('loan_data.csv')
df.head(10)
```

```python
# Age Relationship
sns.boxplot(x='loan_status', y='person_age', data=df)
plt.title('Loan Status by Age')
plt.xlabel('Loan Status')
plt.ylabel('Age')
plt.show()

# Gender Relationship
sns.histplot(x='person_gender', hue='loan_status', data=df, element='step')
plt.title('Loan Status by Gender')
plt.xlabel('Gender')
plt.show()

# Income Relation
sns.violinplot(x='loan_status', y='person_income', data=df)
plt.title('Loan Status by Income')
plt.xlabel('Loan Status')
plt.ylabel('Income')
plt.show()

# Previous Defaults Relationships
sns.histplot(x='previous_loan_defaults_on_file', hue='loan_status', data=df, element='step')
plt.title('Loan Status by Previous Defaults')
plt.xlabel('Previous Defaults')
plt.show()

# Credit Score Relationship
sns.histplot(x='credit_score', hue='loan_status', data=df, element='step')
plt.title('Loan Status by Credit Score')
plt.xlabel('Credit Score')
plt.show()

# Loan Intent Relationship
sns.histplot(x='loan_intent', hue='loan_status', data=df, element='step')
plt.title('Loan Status by Loan Intent')
plt.xlabel('Loan Intent')
plt.xticks(rotation=30, ha="right")
plt.show()

# Home Ownership relationship
sns.histplot(x='person_home_ownership', hue='loan_status', data=df, element='step')
plt.title('Loan Status by Home Ownership')
plt.xlabel('Ownership Type')
plt.show()
```

From the relationships visualized above the ones that are most related to the loan status are the persons age, gender ,credit score, previous defaults, and home ownership on if the loan will be approved or not. The ones that dont seem to relate that well are the persons income and the intent of the loan. I do feel that with this relation the company does discriminate against protected classes based on gender and age with the ammount of rejected loans with higher aged people and with male having a much higher rejection amount than females do.
